```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = np.random.normal(0, 10, (100,1)) # 100 by 1 array
target = np.random.normal(0, 1, 100) # mean 0, 표준편차1, 100개의 samples
```

```python
class LinearReg:
  def __init__(self, data, target, scaling=False):
    self.data= data
    self.target = target
    self.num_instances = self.data.shape[0]
    self.num_features = self.data.shape[1]
    self.scaling = scaling
    print(f"num_instance {self.num_instances}, num_features {self.num_features}, scaling={scaling}")

  def minmax(self): # 0과 1 사이로 정규화
    for i in range(self.num_features):
      col = self.data[:,i]
      self.data[:,i] = (self.data[:,i] - np.min(col))/(np.max(col) - np.min(col))

    return self

  def solver(self):

    if self.scaling == True:
      self.minmax()
    plt.plot(self.data, self.target, ".")
    plt.show()
```

```python
m1 = LinearReg(data, target)
m1.solver()
```

```python
m2 = LinearReg(data, target, True)
m2.solver()
```
