```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
epoch = np.arange(1, 100) # 학습을 n 회 돌렸나
train_loss = 1.5/(epoch) #
val_loss = 0.3+1/(epoch)
```

```python
plt.figure(figsize=(10,5))
plt.plot(train_loss, "r-*") # 변수, 색깔
plt.plot(val_loss, "b-*")
plt.legend(['train','validation']) # plot의 순서대로
plt.title("MNIST")
plt.xlabel('epoch')
plt.ylabel('loss')
```

```python
plt.figure(figsize=(5, 10))
t = np.arange(0., 5., 0.2)
      # x  y 그래프 형태  x y 그래프의 특징(B.lue S.pare) x y 그래프 특징 green ^(삼각형)
plt.plot(t, t, 'ro', t, t**2, 'bs', t, 5*np.sin(t), 'g^')
plt.legend(['t', 't^2', '5sin(t)'])
plt.show()
```

```python
names = ["First", "Business", "Economy"]
prices  = [20, 5, 1]
# 비율
plt.figure(figsize=(15, 5))
plt.subplot(131) # 1열 3행의 첫 번쨰
# 335 -> 3행 3열의 5번
# 1 2 3
# 4 5 6
# 7 8 9
plt.var(names, prices) # 막대형
plt.title(f"{names[0]}", fontsize=14)

plt.subplot(132)
plt.scatter(names, prices) # 점
plt.xlabel('class', fontsize=12)

plt.subplot(133)
plt.plot(names, prices, linewidth = 5.0) # 선
plt.subtitle('Ticket class', fontsize= 14)
plt.axis('off')
plt.show()
```
